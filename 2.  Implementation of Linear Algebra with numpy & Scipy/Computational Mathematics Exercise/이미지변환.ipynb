## <span style='color: blue'>문제</span>

다음 조건을 만족시키는 함수 `f`를 작성하시오. (반드시 함수명을 `f`로)

- 함수 형태: `f(image, color, tol)`

- 입력인수   
>- `image`: RGB 칼라 이미지를 담고 있는 shape가 (h,w,3)인 배열 
>- `color`: RGB 칼라를 각각 0과 1사이의 값으로 표현한 리스트 `[r,g,b]`  
>- `tol`: 두 색의 차이에 대한 기준 값 ($0\le$ `tol` $\le 1$)

- 반환되는 결과
>- `image`와 동일한 shape를 갖는 (h,w,3) 배열 

- 함수 `f`가 하는 일: `image`의 각 픽셀의 색상 `[x,y,z]`와 입력에서 `color`로 주어진 색상 `[r,g,b]`를 각각 비교하여 입력인수 `tol`에 대해   
>- 세 조건 $|x-r|\le \text{tol}$, $|y-g|\le \text{tol}$, $|z-b|\le \text{tol}$을 모두 만족시키면 `[x,y,z]`를 `[r,g,b]`로 대체   
>- 그렇지 않으면 `[x,y,z]`를 `[1-r, 1-g,1-b]`로 대체 


- numpy.ndarray 연산의 장점 활용, nd.array에 대한 불(bool) indexing

```python
def f(image, color, tol):
    color = np.array(color)
    mask = np.all(np.abs(image - color) <= tol, axis=-1)

    result_image = np.where(mask[:,:,None], color, 1-color)
    return result_image
```

```python
face_kcha = plt.imread('kcha.png')
result = f(face_kcha, [0.81, 0.62, 0.53], 0.4)
plt.imshow(result, cmap='brg')
```

```python
# 이 부분을 실행시키시오.

figs, axes = plt.subplots(1,5, sharex=True, sharey=True,figsize=(20,100))
for i in range(5):
    axes[i].imshow(f(face_kcha, [0.81, 0.62, 0.53], (i+1)*0.1),cmap='brg')
```

```python
# 이 부분을 실행시키시오.

figs, axes = plt.subplots(1,5, sharex=True, sharey=True,figsize=(20,100))
for i in range(5):
    axes[i].imshow(f(face_kcha, [0.1, 0.1, 0.1], (i+1)*0.1),cmap='brg')
```
