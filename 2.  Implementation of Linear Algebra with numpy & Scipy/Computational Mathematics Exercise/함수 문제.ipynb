## [문제 1]

$998$ 이하의 음이 아닌 정수들로 이루어진 집합을 $S=\{n\in \mathbb Z: 0\le n \le 998\}$라 하자. 

함수 $f:\mathbb N\to S$가 다음
$
f(n)=\begin{cases}
3 &, n=1\\ 2 &, n=2\\ 1 &, n=3\\ \bigl(f(n-3)-2 f(n-2)+f(n-3)f(n-1)\bigr)\, \rm{mod}\, 999 &, n\ge 4
\end{cases}
$

과 같이 정의될 때, 물음에 답하시오. (단, $m \, \rm{mod}\, 999$는 $m$을 $999$로 나눈 나머지를 의미한다.)

### [문제 1-1]

$f(n)$의 값을 구하는 함수를 작성하고 이를 이용하여 $f(2\times 10^4)$의 값을 구하시오. (10점)

```python
# [문제 1-1] 풀이 코딩 : 아래를 채우시오.

fmemory = {1:3, 2:2, 3:1}
def f1(n):
    if n in fmemory:
        return fmemory[n]
    else:
        temp = (f1(n-3)-2*f1(n-2)+f1(n-3)*f1(n-1))%999
        fmemory[n] = temp
        return temp
```

```python
def f2(n):
    a, b, c = 3, 2, 1
    for i in range(n-2):
        a, b, c = b, c, (a-2*b+a*c)% 999
    return c
```

### [문제 2]

직원이 $N$명인 햄버거 가게에 $M$명의 손님이 도착한다. 이 가게는 문을 열 때 타이머를 동작시켜 문을 닫을 때까지 분 단위로 시간을 재고 있다. 각 손님 $x_m\, (m=1,\cdots,M)$이 $L_m$분에 도착해서 주문을 한 후 음식이 나올 때까지 $K_m$분 걸리는 주문을 할 때, 모든 손님이 음식을 받을 때까지 평균 대기 시간을 구하는 함수 timeToWait()를 작성하시오. (15점)

단, 입력은 $[N,(L_1,K_1),(L_2,K_2),\cdots,(L_M,K_M)]$의 형태로 주어진다. 

예를 들어, 

timeToWait($[2,(2,10),(5,2),(15,4)]$)의 반환값은 0.0이고, 

timeToWait($[1,(2,10),(5,2),(15,4)]$)의 반환값은 2.33이다.

```python
# [문제 2] 풀이 코딩: 아래를 채우시오.

def timeToWait1(iL): #iL : input list     # 음식 수령까지 평균 대기 시간
    n = iL.pop(0) 
    m = len(iL)
    wt_per_line = []  # waiting time per line
    total_wt=0
    timer=0
    iL.sort(key=(lambda x:x[0]))
    lines=[]
    for i in range (0, n):
        lines.append([])
        wt_per_line.append(0)
    while(iL or nf): #nf : not finished
        if (iL and timer==iL[0][0]):
            minidx = wt_per_line.index(min(wt_per_line))
            lines[minidx].append(tuple(iL[0]))
            wt_per_line[minidx]+=iL[0][1]
            iL.pop(0)
        timer+=1
        nf = 0
        for i in range(0, n):
            if (lines[i]):
                lines[i][0] = (lines[i][0][0], lines[i][0][1]-1)
                wt_per_line[i] -=1
                if (lines[i][0][1]==0):
                    total_wt += max(0, timer-lines[i][0][0])
                    lines[i].pop(0)
            if lines[i]:
                nf += 1
    return total_wt/m
```

```python
print(timeToWait1([3,(2,10),(5,2),(15,4)]))
```

```python
def timeToWait2(iL):   #iL: input list    # 주문까지 평균 대기 시간 
    n = iL.pop(0)
    m = len(iL)
    iL.sort(key=(lambda x:x[0]))
    
    wt=0   # waiting time
    clerks = [] 
    for i in range(0, n):
        clerks.append(0)
    
    for i in range(0, m):
        toa, ct= iL.pop(0) #L toa: time of arrival, K ct: cooking time
        idx=clerks.index(min(clerks))
        s=max(clerks[idx], toa)
        wt+=s-toa
        clerks[idx]=s+ct        
    return wt/m
```

```python
print(timeToWait2([1,(2,10),(5,2),(15,4)]))
```

### [문제 3]

다음 Polynomial class는 다항식의 계수를 상수항부터 나열된 리스트(sparse=False경우)로 입력받거나 차수와 계수로 이루어진 딕셔너리 형태(sparse=True인 경우)로 입력받아 다항식과 관련된 연산을 구현하는 것이다. 
(단, 내부적으로 다항식은 차수와 계수로 이루어진 딕셔너리 형태의 self.coeff로 구현한다.)
아래의 Polynomial class를 완성하여 다음 계산이 가능하도록 구현하시오.[15점]

* $f$ = Polynomial({0:2, 1:1, 15:1})
* $g$ = Polynomial([2, -1, 1, 2, 1])
* print($f+g$)
* print($f*g$)
* $h$ = $(2*f+g)**30$
* print($h$)
* print($h$.eval(2))

```python
# [문제 3] 풀이 코딩: 아래를 채우시오.

class Polynomial:
    def __init__(self, coeff=None, sparse=False):
        if not sparse:
            deg_temp = len(coeff)-1
            while not coeff[deg_temp]:
                deg_temp -= 1
            self.__deg = deg_temp
            self.coeff = {i:coeff[i] for i in range(self.__deg+1) if coeff[i]!=0}
        else:
            self.coeff = coeff
            zero_coeff = [key for key in self.coeff if self.coeff[key]==0]
            for key in zero_coeff:
                del self.coeff[key]
            self.__deg = sorted(list(self.coeff.keys()))[-1]
                
    def __add__(self, other):    #a.__add__(b)는 a+b를 구현  (a,b:Polynomial 객체)
        temp = {} 
        for i in (set(self.coeff.keys())| set(other.coeff.keys())):
            temp[i] = self.coeff.get(i,0) + other.coeff.get(i,0)
        return Polynomial(temp, sparse=True)
        
    def __mul__(self, other):    #a.__mul__(b)는 a*b를 구현  (a,b:Polynomial 객체)
        temp = {}
        for i in self.coeff:
            for j in other.coeff:
                temp[i+j] = temp.get(i+j,0)+ (self.coeff[i]*other.coeff[j])
        return Polynomial(temp, sparse=True)
    
    def __rmul__(self, val):     #a.__rmul__(r)은 r*a를 구현 (a:Polynomial 객체, r은 실수)
        temp = {}
        for i in self.coeff:
            temp[i] = val * self.coeff[i]
        return Polynomial(temp, sparse=True)
    
    def __pow__(self, n):      #a.__pow__(n)은 a**n을 구현 (a:Polynomial 객체, n은 자연수)
        res = 1
        remain = n
        power = self
        while remain:
            if remain & 1:
                res = res*power
            power = power * power
            remain >>=1
        return res
    
    def __str__(self):
        temp = []
        for i in sorted(list(self.coeff.keys())):
            if i==0:
                temp.append(f'{self.coeff[i]}')
            elif i==1:
                temp.append(f'{self.coeff[i]}*x')
            else:
                temp.append(f'{self.coeff[i]}*x**{i}')
        return ' + '.join(temp)
    
    def eval(self,val):
        res = 0
        for key in self.coeff:
            res += self.coeff[key]*(val**key)
        return res
    
    @property
    def deg(self):
        return self.__deg
```

```python
# [문제 3] 확인 부분: 위 코드를 작성한 후 아래를 실행시키시오.

f = Polynomial({0:2, 1:1, 15:1},sparse=True)
g = Polynomial([2, -1, 1, 2, 1])

print(f+g)
print()
print(f*g)
print()

h = (2*f+g)**30
print(h)
print()
print(h.eval(2))
print()
```
